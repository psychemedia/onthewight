```python
import time

pclookup={}
for pc in df2['postcode'].unique():
    time.sleep(1)
    print(pc,)
    url='http://uk-postcodes.com/postcode/{pc}.json'.format(pc=pc.replace(' ',''))
    jdata=json.loads(requests.get(url).content)
    pclookup[pc]={'data':jdata,'lat':jdata['geo']['lat'], 'lng':jdata['geo']['lng']}
```

```python
df2['lat']=df2['postcode'].apply(lambda x: pclookup[x]['lat'])
df2['lon']=df2['postcode'].apply(lambda x: pclookup[x]['lng'])
```

```python
#Probably need to install folium
#!pip install folium

#A couple of standard helper routines to preview the maps.

from IPython.display import HTML

import folium
reload(folium)

def mapfix(path):
    f=open(path,'r')
    html=f.read()
    f.close()
    html=html.replace('"//','"http://')
    f=open(path,'w')
    f.write(html)
    f.close()
    
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))

#The embed_map approach saves an HTML file that generates the map if you access if access it via a webserver
#The server stuff is all handled for you when viewing the map in an IPython notebook
def embed_map(map, path="map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    mapfix(path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

```python
def mapper(tmp,df):
    for row in df[ df['lon'].notnull() ].iterrows():
        #print(row[1]['lon'],row[1]['lat'])
        #Define a template that will contain information in the pop-up marker
        template='<strong>{price}</strong><br/>Date: {date}<br/><em>Location:</em> {location}'
        tmp.simple_marker([row[1]['lat'], row[1]['lon']],
                          popup=template.format(price='£{:,}'.format(row[1]['price_paid']),
                                                date=row[1]['deed_date'],
                                                location='{}, {}'.format(row[1]['paon'],row[1]['street'])
                                                ),
                          #This parameter switches on a feature of the leafle map library that auto-clusters numerous close markers
                          clustered_marker=True)
    return tmp
```

```python
#Create a base map - I'm guessing the mid-pointlat-long - we should create it from the data...
longm=df2[ df2['lon'].notnull() ]['lon'].mean()
latm=df2[ df2['lat'].notnull() ]['lat'].mean()

pricesmap = folium.Map(location=[latm, longm], zoom_start=11)
pricesmap=mapper(pricesmap,df2)
#An example of generating the "inline" preview map
inline_map(pricesmap)
```
